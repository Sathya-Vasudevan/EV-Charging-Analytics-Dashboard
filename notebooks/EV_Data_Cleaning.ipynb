```python
# Handling missing values and duplicates 

df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
# Missing Data handled uding group - wise median imputation 

df['Charging Rate (kW)'] = df.groupby('Vehicle Model')['Charging Rate (kW)'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Energy Consumed (kWh)'] = df.groupby('Vehicle Model')['Energy Consumed (kWh)'].transform(lambda x: x.fillna(x.median()))
```

```python
df['Distance Driven (since last charge) (km)'] = df.groupby('Vehicle Model')['Distance Driven (since last charge) (km)'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# Datatype change 

df['Charging Start Time'] = pd.to_datetime(df['Charging Start Time'], dayfirst=True, errors='coerce')
```

```python
df['Charging End Time'] = pd.to_datetime(df['Charging End Time'], dayfirst=True, errors='coerce')
```

```python
df[['Charging Start Time', 'Charging End Time']].head(10)
```

```python
df.info()
```

```python
# Creation of column - Charging Duration

df['Charging Duration'] = (df['Charging End Time'] - df['Charging Start Time']).dt.total_seconds() / 60
```

```python
df[['Charging Duration']].head() 
```

```python
# Fixing outliers / Invalid values 
# 1.Charging Duration 

df = df[df['Charging Duration'] >= 0]
```

```python
# 2.State of Charge (Start %)

df.loc[df['State of Charge (Start %)'] > 100, 'State of Charge (Start %)'] = 100
```

```python
# 3.State of Charge (End %)

df.loc[df['State of Charge (End %)'] > 100, 'State of Charge (End %)'] = 100
```

```python
# Creating Calculate field (for tableau)
# 1.Charging Duration - Already created
# 2.State of Charge (Change %) 

df['State of Charge (Change %)'] = df['State of Charge (End %)'] - df['State of Charge (Start %)']
```

```python
df[['State of Charge (Change %)']].head()
```

```python
# 3. Charging rate (kW)

df['Charging Rate kW'] = df['Energy Consumed (kWh)'] / (df['Charging Duration'] / 60)
```

```python
df.to_csv("ev_charging_patterns_cleaned.csv", index=False)
```
